### Mitä on datatiede?

Datatiede rakentuu neljän kokonaisuuden varaan: liiketoimintaosaaminen, ohjelmointi- ja tietokantaosaaminen, tilastollinen analyysi sekä datalähtöinen viestintä ja visualisointi. Datatieteen prosessia voidaan määritellä esimerkiksi <a href="https://en.wikipedia.org/wiki/Cross-industry_standard_process_for_data_mining">CRISP-DM mallin</a> avulla. Kun ymmäretään mitä käsitellään, esikäsitellään data, luodaan siitä mallinnuksia, mallinnusten perusteella saadaan (toivottavasti) uutta ymmärrystä liiketoiminnasta.

## Datan esikäsittely

Ennen kuin dataa voidaan hyödyntää datatiedeprojektissa, tulee se kerätä haltuun. Tämä voi olla helppoa tai haasteellista, riippuen siitä, millaista dataa ollaan keräämässä. Jos kuvitellaan työelämän organisaatiossa tapahtuvaa datatiedeprojektia, lähdedata on usein peräisin monesta eri järjestelmästä, tehden sen keräämisestä, järjestämisestä ja yhteensovittamisesta haasteellista. Tätä prosessia kutsutaan ETL (Extract, transform, load) prosessiksi, jossa data kerätään eri lähteistä, muunnetaan haluttuun muotoon, ja siirretään datavarastoon, josta sitä voidaan käyttää analysointiin. 

![BI Data](https://www.altexsoft.com/media/2019/04/etl.png)

Internetistä dataa voidaan kerätä ryömijöiden ja raapijoiden avulla. Ryömijä (web crawler) on robotti, joka käy systemaattisesti läpi verkko-osoitteita. Sen tehtävänä on kerätä tietoa verkkosivuilta ja indeksoida se esim. hakukoneita varten. Ryömijöitä käytetään monissa eri tarkoituksissa, kuten hakukoneoptimoinnissa, tietojen keräämisessä, sisällön indeksoinnissa ja vertailussa sekä verkkosivujen seurannassa.

Raapija (web scraper) puolestaan kerää automatisoidusti tietoa verkkosivuilta. Sen pääasiallinen tehtävä on hakea ja tallentaa tietoa verkkosivuilta tietokantaan tai tiedostoon. Raapijat toimivat suoraan verkkosivun HTML rakenteen kanssa. Raapijoita käytetään laajasti erilaisissa sovelluksissa ja tarkoituksissa, kuten tiedon keräämisessä, kilpailijoiden seurannassa, hintojen vertailussa, markkinatutkimuksessa ja sisällön analysoinnissa.

Python-kirjastoja verkkosivujen raapimiseen ovat esimerkiksi <a href="https://www.crummy.com/software/BeautifulSoup/">Beautiful Soup</a> ja <a href="https://scrapy.org/">Scrapy</a>.

Ryömijöiden ja raapijoiden käyttö ei ole ongelmatonta. <a href="https://benbernardblog.com/web-scraping-and-crawling-are-perfectly-legal-right/">Artikkeli ryömijöiden ja raapijoiden laillisuudesta</a> paljastaa, että niiden hyödyntäminen voi olla lain vastaista. Kaiken datan (vaikka se olisi julkisesti saatavilla) omistaa joku, ja data saattaa olla titetosuojattua. Ryömijöiden/raapijoiden käyttö saattaa myös olla palveluiden käyttöoikeuksien vastaista. On myös mahdollista, että esim. raapimalla ihmisten henkilökohtaisiin asioihin liittyvää dataa rikot EU:n GDPR asetuksia. Kannattaa siis olla tarkkana. 

Tässä työkirjassa dataan päästään helposti käsiksi lataamalla se suoraan <a href="https://insideairbnb.com/get-the-data/">Inside Airbnb-palvelusta</a>. Tämän jälkeen dataa pitää kuitenkin esikäsitellä. Datassa saattaa olla paljon esim. tyhjiä arvoja, turhia sarakkeita tai nolla-arvoja, jotka halutaan poistaa. Lisäksi voidaan esim. siistiä sarakkeiden tai tietueiden nimiä, tai luoda kokonaan uusia muuttujia, jos huomataan sarakkeiden tietueista löytyvän tietoa, jotka kannattaa jakaa useampaan muuttujaan.

Pandas-kirjasto on erinomainen väline datan jalostamiseen. Katsotaan seuraavaksi esimerkki siitä, miten työkirjassa käytettyä esimerkkidataa Amsterdamin AirBnB asunnoista voisi siivota, muokata sekä jalostaa. 


```python
# Tuodaan pandas-kirjasto datan lukemista, hallintaa ja käsittelyä varten 
import pandas as pd
```

```python
# Tarkastellaan mllaista data on. Data on luettavissa suoraan InsideAirBnB:n nettisivuilta
url = 'https://data.insideairbnb.com/the-netherlands/north-holland/amsterdam/2023-12-12/visualisations/listings.csv'
# Luodaan dataframe hyödyntäen Pandas kirjastoa
df = pd.read_csv(url)

#(Mikäli suoraan nettisivuilta lataaminen ei toimi,
#olen sisällyttänyt datan myös datatiedeprojekti-kansioon)
# df = pd.read_csv('./data/Amsterdam_listings.csv')

# Luetaan 5 ensimmäistä tietuetta
df.head()
```

Huomataan, että datassa on paljon hyödyllistä tietoa, mutta myös paljon turhaa, siispä sitä pitää siivota. Ennen siivoamista katsotaan kuitenkin datan ominaispiirteitä. Katsotaan aluksi vaikka, ovatko kaikki ID-arvot yksikäsitteisiä. Näin saadaan selville, onko datan seassa duplikaatteja.

```python
#Ovatko ID-arvot yksikäsitteisiä
df['id'].is_unique
```

Saadaan selville, että datassa ei ole duplikaatteja. ID:n voisi asettaa indeksiksi, mutta huomasin sen olevan vähän kimuranatti muuttuja myöhemmissä vaiheissa, joten poistan sen. 

```python
#Poistetaan ID
df = df.drop(['id'], axis=1)
df.head()
```

Haetaan seuraavaksi lisätietoja datasta, jotta saadaan selville mitä muuta datalle kannattaisi tehdä. 

```python
#Haetaan datasta lisätietoa 
print("Datan koko:", df.shape)
print("Datan tyypit:", df.dtypes)
print("Puuttuvien arvojen määrä:")
print(df.isnull().sum())
print("Yksilöllisten arvojen määrä:")
df.select_dtypes(include=['object']).nunique()
```

Huomataan esimerkiksi, että `neighbourhood_group` -arvo puuttuu kaikista kohteista, tämä voidaan siis huoletta poistaa kokonaan. Voidaan myös miettiä, että mitkä arvoista ovat tämän datatiedeprojektin kannalta hyödyttömiä? En usko, että `host_id`, `host_name`, `last_review` tai `license` -kentä tuovat analyysiin mitään lisäarvoa, joten poistetaan myös ne. 

```python
#Poista tarpeettomat sarakkeet
df = df.drop(['host_id', 'host_name', 'neighbourhood_group', 'last_review', 'license'], axis=1)
df.head()
```

Voidaan huomata myös, että `rating`-arvo on piilossa `name` sarakkeessa, kuten myös huoneiden määrä. Name sarakkeessa on muutenkin tietoa aivan liian paljon tietoa, tehden datasta epäselvän näköistä. Siistitään siis `name`-saraketta sekä erotetaan `rating` arvo sekä huoneiden määrä sieltä omiksi sarakkeekseen. 

```python
# Erotetaan rating ja huoneiden määrä omiksi sarakkeiksi
df[['rating', 'bedrooms']] = df['name'].str.extract(r'★([\d.]+).*?(\d+)\s*bedroom')

# Rating ja bedrooms numeroksi
df['rating'] = pd.to_numeric(df['rating'])
df['bedrooms'] = pd.to_numeric(df['bedrooms'])

#Siivotaan nimeä niin, että vain alkuosa jää
df['name'] = df['name'].str.split('·').str[0]

#Tuodaan vielä sarakkeet nimen viereen. Muokataan muutenkin järjestystä hieman paremmaksi 
column_order = ['name', 'bedrooms', 'rating', 'price', 'room_type','neighbourhood','minimum_nights', 'number_of_reviews', 'reviews_per_month','number_of_reviews_ltm','availability_365', 'latitude', 'longitude', 'calculated_host_listings_count']
df = df[column_order]

df.head()
```

Data alkaa näyttää jo siistimmältä. Huomataan kuitenkin, että `availability_365` sarakkeessa on 0-arvoja. Muokkauksen jälkeen katsotaan uudestaan puuttuvien arvojen määrää. Katsotaan samalla, onko datassa paljonkin 0-arvoja.  

```python
# Laskee, kuinka monta arvoa kussakin sarakkeessa on 0
zero_counts = (df == 0).sum(axis=0)
print("0-arvojen määrä sarakkeittain:")
print(zero_counts)

print("Puuttuvien arvojen määrä:")
print(df.isnull().sum())
print("Datan kokonaismäärä:")
df.shape
```

Huomataan, että melko suuri osa datasetistä olikin nimen osalta ilmeisesti tehty eri tavalla, ja tehdyllä erottelulla 2173 kohtaa jäävät ilman arvoa. Huoneiden lukumäärä ja arvostelu ovat kuitenkin analyysin kannalta kiinnostavia arvoja, joten tehdään tässä vaiheessa päätös, että poistetaan datasetistä puuttuvat arvot. Myös `reviews_per_month` -sarakkeessa on runsaasti tyhjiä arvoja. Koska ei ole mitään varmuutta siitä, että ovatko nämä sellaisia tietueita joiden kohdalta puuttuisi arvot myös `bedrooms` ja `rating` sarakkeista, poistetaan koko `reviews_per_month` sarake, sillä se on paljon epäolennaisempaa tietoa. Hyödynnetään siinä <a href="https://builtin.com/data-science/regression-machine-learning"> regressiomalleja.</a> 



```python
#Tuodaan koneoppimisen kirjastoja
import sklearn
from sklearn import linear_model
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
#Tarkastellaan dataa vielä kerran
print(df.dtypes)
print("Puuttuvien arvojen määrä:")
print(df.isnull().sum())
```

Puuttuvia arvoja eie ole mistään ilmestynyt maagisesti lisää. Kategoriset muuttujat pitää muuttaa numeerisiksi koneoppimista varten, tai vaihtoehtoisesti poistaa. Tässä tapauksessa sekä naapurusto että huonetyyppi ovat merkittäviä tekijöitä asuntojen hintaa ajatellen, joten muutetaan kategoriset muuttujat numeerisiksi.

Asetetaan nimi indeksiksi, jotta sitä ei tarvitse muuttaa kategoriseeksi muuttujaksi/poistaa. Tämän jälkeen jaetaan data ominaisuuksiin ja kohdemuuttujaan. Seuraavaksi data jaetaan training ja testidatoihin, joiden avulla koitetaan ennustaa hintoja. 

Data voidaan vielä normalisoida hyödyntäen esimerkiksi scikit-learnin <a href="https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html">standard scaleria.</a> 

```python
# Asetetaan nimi indeksiksi muuttujan sijasta
df.set_index('name', inplace=True)

#Poistetaan hinta-sarakkeesta ääriarvoja 
df = df[df['price'] <= 600]

# Kategoristen muuttujien listaaminen
categorical_cols = ['room_type', 'neighbourhood']

# One-hot encoding 
ml_df = pd.get_dummies(df, columns=categorical_cols)

# Jaetaan data ominaisuuksiin (X) ja kohdemuuttujaan (y)
X = ml_df.drop('price', axis=1)
y = ml_df['price']

# Jaetaan datasetti kahteen osaan: training- ja testing-dataan. Otetaan testausdataksi 50% training datasta
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Skaalataan data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Luo kopio testidatataulukosta
predicted_df = X_test.copy()  

```

```python
# Testataan seuraavia lineaarisen regression malleja 
rs = 1
ests = [ linear_model.LinearRegression(), linear_model.Ridge(),
        linear_model.Lasso(), linear_model.ElasticNet(),
        linear_model.BayesianRidge(), linear_model.OrthogonalMatchingPursuit() ]
ests_labels = np.array(['Linear', 'Ridge', 'Lasso', 'ElasticNet', 'BayesRidge', 'OMP'])
errvals = np.array([])

#For-silmukassa kullekin regressiomallille suoraitetaan sovitus skaalattuun dataan ja kohdedetaan
for e in ests:
    e.fit(X_train_scaled, y_train)
    #Lasketaan kunkin mallin ennustama keskimääräinen virhe 
    this_err = metrics.median_absolute_error(y_test, e.predict(X_test_scaled))
    #Ennustetut hinnat tallennetaan predicted prices-muuttujaan ja viedään predicted_df dataframeen jatkotarkastelua varten 
    predicted_prices = e.predict(X_test_scaled)
    predicted_df['predicted_price'] = predicted_prices
    print(f"got error {this_err:.2f}")
    #Tallennetaan kunkin regressiomallin keskimääräinen virhe 
    errvals = np.append(errvals, this_err)
    pos = np.arange(errvals.shape[0])

#Tarkastellaan MAE:ta (Median absolute error) eri malleilla    
srt = np.argsort(errvals)
plt.figure(figsize=(7,5))
plt.bar(pos, errvals[srt], align='center')
plt.xticks(pos, ests_labels[srt])
plt.xlabel('Estimator')
plt.ylabel('Median Absolute Error')                  
```

Malli toimii, mutta huomataan, että virhe on suhteellisen suuri. Tarkastellaan virhettä visuaalisemmassa muodossa, jotta voidaan havainnollistaa paremmin, mistä se voisi johtua:

```python
#Tarkastellaan, kuinka hyvin hintaennustinmalli toimii histogrammin avulla vertailemalla todellisiin hintoihin
plt.figure(figsize=(10, 6))
plt.hist(y_test, bins=30, alpha=0.5, color='blue', label='Todelliset hinnat')
plt.hist(predicted_prices, bins=30, alpha=0.5, color='red', label='Ennustetut hinnat')
plt.xlabel('Hinta')
plt.ylabel('Frekvenssi')
plt.title('Todelliset hinnat vs. Ennustetut hinnat')
plt.legend()
plt.grid(True)
plt.show()
```

Kuten kuvaajasta nähdään, hintojen ennustajamalli ei missään nimessä ole täydellinen. Keskivirhettä on aika paljon, ja se tuntuu ennustavan paljon täysin samanlaisia hintoja. Nopealla katsauksella en löytänyt tähän ongelmaan vastausta, vaan se tarvitsisi tarkempaa paneutumista. Dataa on huomattavan suuri määrä, ja kun kokeilin analysoida kaikkea, siinä kesti tuhottoman kauan. Poistetaan tästä syystä puuttuvat arvot ja otetaan datasta otos, jota analysoidaan.

```python
#Poistetaan puuttuvat arvot
df_reviews = df_reviews.dropna(axis=0)

#Otetaan datasta otos
sample_size = 25000
sample_df = df_reviews.sample(n=sample_size, random_state=42)  
```

Tehdään seuraavaksi hieman lisää esikäsittelyä kommenteille ja koitetaan löytää 20 yleisimmin käytettyä sanaa, jonka jälkeen tehdään kuvaaja

```python
# Poistetaan englanninkieliset "stopwordit" ja tehdään translator, jotta voidaan poistaa välimerkit
stop_words = set(stopwords.words('english'))
translator = str.maketrans('', '', string.punctuation)

# Prosessoidaan kommentit
comment_cleaned = []
for comment in sample_df['comments']:
    # Poistetaan välimerkit
    comment = comment.translate(translator)
    # Tokenisaatio
    tokens = word_tokenize(comment)
    # Poistetaan sanat joissa on numero
    tokens = [token for token in tokens if not token.isdigit()]
    # Pidetään vain yli 3 kirjaimiset sanat
    tokens = [token for token in tokens if len(token) > 3]
    # Poistetaan "stopwordit"
    tokens = [token for token in tokens if token.lower() not in stop_words]
    comment_cleaned.extend(tokens)
        
# Lasketaan sanojen esiintymiskerrat
word_counts = Counter(comment_cleaned)

# Yleisimmät sanat ja niiden esiintymiskerrat
most_common_words = word_counts.most_common(20)
words, frequencies = zip(*most_common_words)

# Tehdään pylväskaavio
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies, color='skyblue')
plt.ylabel('Esiintymiskerrat')
plt.title('20 yleisintä sanaa')
plt.xticks(rotation=60)
plt.show()
```

Huomataan arvostelujen yleisilmeen olevan varsin positiivinen. Tämä toki ei yllätä ottaen huomioon AirBnB:n luonteen arvostelujen suhteen.

### Aihemallinnus 

Katsotaan aihemallinnusta esimerkin kautta. Mitä muuta AirBnB arvosteluista voisi selvittää? <a href="https://www.cs.columbia.edu/~blei/topicmodeling.html">Aihemallin</a> luomisella voidaan selvittää dokumentissä esiintyviä abstrakteja "aiheita". Hyödynnetään tässä <a href="https://en.wikipedia.org/wiki/Latent_Dirichlet_allocation">LDA-mallia</a>  

Käytetään aihemallin luomisessa<a href="https://radimrehurek.com/gensim/">Gensim kirjastoa</a>.

```python
#Tuodaan Gensim kirjasto 
from gensim.models import Phrases
from gensim.models import Word2Vec
from gensim import corpora, models, similarities, matutils
from gensim.models import hdpmodel, ldamodel, TfidfModel
```

```python

# Luodaan dictionary ja korpus
dictionary = corpora.Dictionary(line.lower().split() for line in comment_cleaned)
limit = 1
once_ids = [tokenid for tokenid, docfreq in dictionary.dfs.items() if docfreq <= limit]
# poistetaan sanoja, jotka esiintyvät vain kerran
dictionary.filter_tokens(once_ids) 
#poistetaan välejä 
dictionary.compactify() 
dictionary.save("corpora.dict")
corpus = [dictionary.doc2bow(line.lower().split()) for line in comment_cleaned]

# LDA-mallin koulutus
num_topics = 4  # Montako aihetta halutaan 
lda_model = models.LdaModel(corpus, id2word=dictionary, num_topics=num_topics)

# Printataan aiheet ulos
for topic_id, topic_words in lda_model.show_topics(num_topics=num_topics, formatted=False):
    print(f"Topic {topic_id + 1}:")
    print([word for word, _ in topic_words])
    
```

Huomataan, että datasta saadut aihemallit ovat myös hyvin positiivisia. Niistä piirtyy kuva Amsterdamista kauniina kaupunkina, jossa on hyvä kävellä, sekä paljon ystävällisiä ihmisiä, keskusta on nostettu esiin. Asuntoja kuvaillaan siisteiksi,"täydellisiksi", ihaniksi ja mukaviksi, hyvät sijainnit ovat nostettu myös esiin.

Näistähän voisi tehdä vaikka markkinointimateriaalia Amsterdamille wordcloudin muodossa!

```python
# Tuodaan wordcloud-kirjasto 
from wordcloud import WordCloud

# Tehdään aihemallin sanoista lista
all_words = [word for topic in lda_model.show_topics(num_topics=num_topics, formatted=False) for word, _ in topic[1]]

# Lasketaan sanojen frekvenssi
word_freq = {word: all_words.count(word) for word in set(all_words)}

# Tehdään wordcloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Visualisoidaan se matplotlibillä 
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Amsterdam Word Cloud')
plt.show()
```
